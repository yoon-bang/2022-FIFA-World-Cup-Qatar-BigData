## Dataset과 Player23 합치기

```python
player_features = df_player.keys().to_list()[1:-2]
player_features.pop(1)
player_features
```

```python
# value 

for j in player_features:
  # 수비수 4명
  for i in range(1, 5):
    col = 'B'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 미드필더 4명 
  for i in range(1, 5):
    col = 'M'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 공격수 2명 
  for i in range(1, 3):
    col = 'T'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 골키퍼 1명 
  col = 'K1_'+str(j)
  dataset[col] = np.nan
```

```python
for index in notebook.tqdm(dataset.index):
  team1 = dataset.loc[index, ['team']]
  team1 = team1[0]

  # position B (수비수) 4명 
  for i in range(1, 5):
    for j in player_features:
      col1 = 'B'+str(i)+'_'+j
      try:
        data = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'B') & (df_player['nationality_name'] == team1 )]
        dataset.loc[index, [col1]] = float(data)
      except:
        continue

  # position M (미드필더) 4명 
  for i in range(1, 5):
    for j in player_features:
      col1 = 'M'+str(i)+'_'+j
      try:
          data = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'M') & (df_player['nationality_name'] == team1 )]
          dataset.loc[index, [col1]] = float(data)
      except:
        continue

  # position (공격수) 2명 -> T
  for i in range(1, 3):
    for j in player_features:
      col1 = 'T'+str(i)+'_'+j
      try:
          data = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'T') & (df_player['nationality_name'] == team1 )]
          dataset.loc[index, [col1]] = float(data)
      except:
        continue

  # # 골키퍼 
  for j in player_features:
    col1 = 'K1_'+j
    try:
        data = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'T') & (df_player['nationality_name'] == team1 )]
        dataset.loc[index, [col1]] = float(data)
    except:
      continue
```
