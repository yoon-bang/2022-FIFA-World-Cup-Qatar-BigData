```python
for index in notebook.tqdm(dataset.index):
  team1 = dataset.loc[index, ['home_team']]
  team2 = dataset.loc[index, ['away_team']]
  date = dataset.loc[index, ['date']] # Modified_for_latest_ranking
  date = str(date[0])[:10]  # Modified_for_latest_ranking
  team1 = team1[0]
  team2 = team2[0]
  try:
    # Modified_for_latest_ranking
    # data1 = df_rankings['total_points'].loc[(df_rankings['country_full'] == team1) & (df_rankings['year'] == year)]
    data1 = df_rankings['total_points'].loc[(df_rankings['country_full'] == team1) & (df_rankings['rank_date'] <= date)].to_frame().iloc[0]
    # data2 = df_rankings['total_points'].loc[(df_rankings['country_full'] == team2) & (df_rankings['year'] == year)]
    data2 = df_rankings['total_points'].loc[(df_rankings['country_full'] == team2) & (df_rankings['rank_date'] <= date)].to_frame().iloc[0]

    dataset.loc[index, ['home_team_total_points']] = float(data1)
    dataset.loc[index, ['away_team_total_points']] = float(data2)
  except:
        continue
```

```python
df_player14['year'] = 2014
df_player15['year'] = 2015
df_player16['year'] = 2016
df_player17['year'] = 2017
df_player18['year'] = 2018
df_player19['year'] = 2019
df_player20['year'] = 2020
df_player21['year'] = 2021
df_player22['year'] = 2022
```

```python
df_player = pd.concat([df_player14, df_player15, df_player16, df_player17, df_player18, df_player19, df_player20, df_player21, df_player22], ignore_index=True)
```

```python
# nationality_name, Model_Position, year 로 group by 한 후, Position_Best_Rating 에 대해 rank 매기기 
df_player['rank_in_nation'] = df_player.groupby(['nationality_name','Model_Position', 'year'])['Position_Best_Rating'].rank(method='first',ascending=False)
```

## Datset과 Player 데이터 합치기

```python
player_features = df_player.keys().to_list()[1:-3]
player_features.pop(1)
player_features
```

```python
# home team 
for j in player_features:
  # 수비수 4명
  for i in range(1, 5):
    col = 'home_B'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 미드필더 4명 
  for i in range(1, 5):
    col = 'home_M'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 공격수 2명 
  for i in range(1, 3):
    col = 'home_T'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 골키퍼 1명 
  col = 'home_K1_'+str(j)
  dataset[col] = np.nan
```

```python
# away team 
for j in player_features:
  # 수비수 4명
  for i in range(1, 5):
    col = 'away_B'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 미드필더 4명 
  for i in range(1, 5):
    col = 'away_M'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 공격수 2명 
  for i in range(1, 3):
    col = 'away_T'+str(i)+'_'+str(j)
    dataset[col] = np.nan

  # 골키퍼 1명 
  col = 'away_K1_'+str(j)
  dataset[col] = np.nan
```

```python
for index in notebook.tqdm(dataset.index):
  team1 = dataset.loc[index, ['home_team']]
  team2 = dataset.loc[index, ['away_team']]
  
  year = dataset.loc[index, ['year']]
  year = int(year)

  team1 = team1[0]
  team2 = team2[0]

  # position B (수비수) 4명 
  for i in range(1, 5):
    for j in player_features:
      col_str1 = 'home_B'+str(i)+'_'+str(j)
      col_str2 = 'away_B'+str(i)+'_'+str(j)
      try:
        temp1 = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'B') & (df_player['nationality_name'] == team1 ) & (df_player['year'] == year)]
        temp2 = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'B') & (df_player['nationality_name'] == team2 ) & (df_player['year'] == year)]
        dataset.loc[index, [col_str1]] = float(temp1)
        dataset.loc[index, [col_str2]] = float(temp2)
      except:
        continue

  # position M (미드필더) 4명 
  for i in range(1, 5):
    for j in player_features:
      col_str1 = 'home_M'+str(i)+'_'+str(j)
      col_str2 = 'away_M'+str(i)+'_'+str(j)
      try:
        temp1 = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'M') & (df_player['nationality_name'] == team1 ) & (df_player['year'] == year)]
        temp2 = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'M') & (df_player['nationality_name'] == team2 ) & (df_player['year'] == year)]
        dataset.loc[index, [col_str1]] = float(temp1)
        dataset.loc[index, [col_str2]] = float(temp2)
      except:
        continue
        
  # position (공격수) 2명 -> T
  for i in range(1, 3):
    for j in player_features:
      col_str1 = 'home_T'+str(i)+'_'+str(j)
      col_str2 = 'away_T'+str(i)+'_'+str(j)
      try:
        temp1 = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'T') & (df_player['nationality_name'] == team1 ) & (df_player['year'] == year)]
        temp2 = df_player[str(j)].loc[(df_player['rank_in_nation'] == i) & (df_player['Model_Position'] == 'T') & (df_player['nationality_name'] == team2 ) & (df_player['year'] == year)]
        dataset.loc[index, [col_str1]] = float(temp1)
        dataset.loc[index, [col_str2]] = float(temp2)
      except:
        continue

  # # 골키퍼 
  for j in player_features:
    col_str1 = 'home_K1_'+str(j)
    col_str2 = 'away_K1_'+str(j)
    try:
      temp1 = df_player[str(j)].loc[(df_player['rank_in_nation'] == 1) & (df_player['Model_Position'] == 'K') & (df_player['nationality_name'] == team1 ) & (df_player['year'] == year)]
      temp2 = df_player[str(j)].loc[(df_player['rank_in_nation'] == 1) & (df_player['Model_Position'] == 'K') & (df_player['nationality_name'] == team2 ) & (df_player['year'] == year)]
      dataset.loc[index, [col_str1]] = float(temp1)
      dataset.loc[index, [col_str2]] = float(temp2)
    except:
      continue
```
